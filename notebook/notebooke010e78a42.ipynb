```python
import os
from ultralytics import YOLO

# Set the environment variable to skip W&B
os.environ['WANDB_MODE'] = 'disabled'

model = YOLO('yolo11x-seg.pt')
model.train(
    data='/kaggle/working/breast-cancer-v2-1/data.yaml',
    epochs=100,
    batch=16,
    device=[0, 1],
    seed=42,
    patience=30,
    verbose=True,
    optimizer='SGD',
    lr0=0.0001,
    lrf=0.1,
    momentum=0.9,
    cos_lr=True
)
```

```python
model = YOLO('/kaggle/working/runs/segment/train/weights/best.pt')
metrics = model.val()
metrics.box.map
```
